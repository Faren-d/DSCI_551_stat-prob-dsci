```python
import numpy as np
from numpy.random import randn, rand, seed
import matplotlib.pyplot as plt
seed(7)
```

### 1. Vectors

A vector is a container that holds multiple numbers, and it is written as:

$$x = \begin{pmatrix}x_1\\x_2\\ \vdots \\ x_d \end{pmatrix}.$$

For example, $$x = \begin{pmatrix}2\\-1 \\ 5.5 \end{pmatrix}$$ is a 3-dimensional vector because it has 3 elements. This is like the math equivalent of a list or array in `R` or `Python`, but the values will be numerical. To say $x$ is a $d$-dimensional vector, we can write $x\in \mathbb{R}^d$. Moreover, you can think of a vector as a point in $d$-dimensional space. We will create $n = 5$ random vectors of dimension $d = 2$. The plot below shows them as points with the value of $x$ on the horizontal axis and $y$ on the vertical axis.

```python
n = 5
x, y = randn(n),randn(n)
plt.scatter(x, y);
plt.xlim(-2, 2);
plt.ylim(-2, 2);
plt.show()
```

We can also draw them as arrows starting from the **origin** $(0, 0)$:

```python
plt.quiver(0 * x, 0 * y, x, y, angles = "xy", scale_units = "xy", scale = 1)
plt.scatter(x, y)
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.show()
```

We can also draw them as arrows starting from the **origin** $(0, 0)$:

```python
plt.quiver(0 * x, 0 * y, x, y, angles = "xy", scale_units = "xy", scale = 1)
plt.scatter(x, y)
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.show()
```

### 2. Matrices

A matrix is a 2D container that holds numbers. Nonetheless, the word **dimension** has 2 meanings:

- We refer to the length of a vector as its dimension, because we think of it as a point in $d$-dimensional space,- but in terms of being a container holding numbers, it is a $1$-dimensional container regardless of its length

```python
x = randn(5)
x
```

We can access to the first element of vector `x` as follows:

```python
x[0]
```

We define the above `x` as a 5-dimensional vector because it is a point in $5$-dimensional space:

```python
x.shape
```

But, it is also 1-dimensional:

```python
x.ndim
```

It would be less confusing to call it a **"vector of length 5"** rather than **"a vector of dimension 5"**, but this is how it is done in practice. Thus, it is important to be aware of this matter.

Now, when it comes to matrices, their dimensions refer to the number of rows and columns. So, for instance, the example below `y` is a $3 \times 3$ matrix.

```python
y = randn(3, 3)
y
```

We can access to the element in the second column and first row as follows:

```python
y[2,1]
```

The number of rows and columns of `y` are given as follows:

```python
y.shape
```

Note matrices do not have to be **square** (i.e., same number of rows and columns). For instance, this is a matrix with 3 rows and 5 columns (i.e., a $3 \times 5$ matrix):

```python
x = rand(3,5)
x
```

```python
x.shape
```

We will only deal with **1D containers** (**vectors**) and **2D containers** (**matrices**). These are called **tensors** in Machine Learning jargon. We do not touch 3D (or higher) containers. However, we will deal with vectors and matrices of various sizes in MDS! 

### 3. Random Vectors

In addition to a vector containing numbers, you can also have a vector containing random variables. This is called a **random vector**. We could refer to it as a multidimensional random variable.
This is convenient for representing systems with many random variables, so we do not need to write$$X, Y, Z, A, B, C, \ldots$$

We can just write $V$ as a $d$-dimensional vector, and its elements are the random variables. For instance:$$V = \begin{pmatrix}X \\ Y \\ Z \\ A \\ B \\ C \end{pmatrix}.$$
